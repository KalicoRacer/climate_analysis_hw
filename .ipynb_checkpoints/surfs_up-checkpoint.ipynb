# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Calculate the date 1 year ago from the last data point in the database
last_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()[0]
date_yr_ago = datetime.strptime(last_date, '%Y-%m-%d') - timedelta(days=365)

print(f"The last date in the DB is: {last_date}\nThe date 1 year from that date is: {date_yr_ago}")
```

```python
# Perform a query to retrieve the data and precipitation scores

# Save the query results as a Pandas DataFrame and set the index to the date column

# Sort the dataframe by date

select = [Measurement.date,
         Measurement.prcp]

prec_df = pd.read_sql(session.query(*select).\
                 filter(Measurement.date >= date_yr_ago).\
                 statement,session.bind)

prec_df.sort_values(by='date', ascending=False, inplace=True)
prec_df.dropna(inplace=True)
# prec_df.set_index('date', inplace=True)

prec_df.head()
```

```python
# Use Pandas Plotting with Matplotlib to plot the data

plt.figure(figsize=(15,10))
plt.xticks([])
plt.xlabel('Date')

plt.plot(prec_df['date'], prec_df['prcp'])

plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data

prec_df.describe()
```

```python
# Design a query to show how many stations are available in this dataset?

# SELECT COUNT(id) FROM station
station_count = pd.read_sql("SELECT COUNT(id) AS num_of_stations FROM station",\
                            session.bind)

# station_count = pd.read_sql(session.query(*select).\
#                  statement,session.bind)
x = station_count.iloc[0]['num_of_stations']
print(f'The number of weather stations is: {x}')
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.

station_activity = pd.read_sql("SELECT station AS station_id, COUNT(station) AS num_of_measurements \
                                FROM measurement WHERE prcp IS NOT null \
                                GROUP BY station ORDER BY num_of_measurements DESC",\
                               session.bind)
station_activity
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?

temperature_measurement = pd.read_sql("SELECT station AS station_id, COUNT(station) AS num_of_measurements, \
                                MIN(tobs) AS lowest_temp, MAX(tobs) AS highest_temp, AVG(tobs) AS avg_temp \
                                FROM measurement WHERE prcp IS NOT null\
                                GROUP BY station ORDER BY num_of_measurements DESC",\
                               session.bind)
temperature_measurement
```

```python
# Choose the station with the highest number of temperature observations.

station_temp_obsv = pd.read_sql("SELECT station AS station_id, COUNT(tobs) AS num_of_observations \
                                FROM measurement GROUP BY station ORDER BY num_of_observations DESC;",\
                               session.bind)
station_temp_obsv
```

```python
station_x = station_temp_obsv.iloc[0][0]
station_x
```

```python
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

select = [Measurement.date,
         Measurement.tobs]

temp_df = pd.read_sql(session.query(*select).\
                 filter(Measurement.date >= date_yr_ago, Measurement.station == station_x).\
                 statement,session.bind)

temp_df.sort_values(by='date', ascending=False)

print(len(temp_df))
temp_df.head()
```

```python
plt.figure(figsize=(15,10))
plt.xticks([])
plt.xlabel('Date')

plt.plot(temp_df['date'], temp_df['tobs'])

plt.show()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.

vacation_start = '2017-06-01'
vacation_end = '2017-06-15'
vacation_temps = calc_temps(vacation_start, vacation_end)
vacation_temps
```

```python
avg_temp = vacation_temps[0][1]
avg_temp
```

```python
# x_axis = np.arange(len(users))

plt.figure(figsize=(6,10))
plt.xticks([])
plt.xlim(-1, 1)
plt.ylim(0, 100)

plt.bar(vacation_start, avg_temp, color='coral', alpha=0.5, align="center", width=1.1)

plt.title("Trip Avg Temp", fontsize = 30)
plt.ylabel("Temp (F)", fontsize = 25)
```
